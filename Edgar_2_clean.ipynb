```python
response = open('data/424B2_d795231d424b2.htm')
html_doc = response.read()
soup = BeautifulSoup(html_doc, 'html.parser')
file = soup.prettify()
text= soup.get_text()
text = soup.get_text()
text = unicodedata.normalize('NFKD', text)
text = re.sub('[\s]*\n[\s]*','\n', text)
text = re.sub(' +',' ', text)
text = text.strip()
text = text.lower()
```
